```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import solve_ivp
import sympy
from qbee import *
```

$$
\begin{cases}
y' = -2y+2ay^{2}\\
x' = -x + axy
\end{cases}
$$

```python
plt.style.use('seaborn-poster')
%matplotlib inline

a = 5.0
b = 1.0
F = lambda t, y: -2*y + 2*a*y**2
F2 = lambda t, y: -2*y + 2*b*y**2 
```

```python
t_eval = np.linspace(0, 10, 1000)
sol = solve_ivp(F, [0, 10], [0.1], t_eval=t_eval)
sol2 = solve_ivp(F2, [0, 10], [0.1], t_eval=t_eval)

plt.figure(figsize=(10, 6))
plt.plot(sol.t, sol.y[0], 'b', linewidth=2, label='ODE solution 1')
plt.plot(sol2.t, sol2.y[0], 'r', linewidth=2, label='ODE solution 2')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 定义常数a
a = 1

t = np.linspace(0, 10, 1000)
dt = t[1] - t[0]
x0_1 = 0.5
x0_2 = -0.5
x_values_1 = []
x_values_2 = []
x_1 = x0_1
x_2 = x0_2
for i in range(len(t)):
    x_values_1.append(x_1)
    x_1 += (-x_1 + a * x_1 ** 3) * dt
    
    x_values_2.append(x_2)
    x_2 += (-x_2 + a * x_2 ** 3) * dt

# plot the graph of x against t, with different initial values
plt.figure(figsize=(10, 7))
plt.plot(t, x_values_1, label='x0 = 0.5', color='tab:blue', linewidth=2)
plt.plot(t, x_values_2, label='x0 = -0.5', color='tab:red', linewidth=2)
plt.xlabel('t', fontsize=15)
plt.ylabel('x(t)', fontsize=15)
plt.title(r'$x^{\prime} = -x + ax^3$', fontsize=20)
plt.legend(fontsize=15)
plt.grid(alpha=0.3)
plt.show()
```
