```python
import numpy as np
import matplotlib.pyplot as plt
```

$\left|\Re\left(\lambda_1\right)\right|=\left|\max(-c, \frac{b+\sqrt{b^{2}+4k}}{2})\right|$ and $\|F_{2}\|=\max(\left|a\right|,\sqrt{c^{2}+4})$. Then we got
$$
R := \frac{\|X_{0}\|\max(\left|a\right|,\sqrt{c^{2}+4})}{\left|\max(-c, \frac{b+\sqrt{b^{2}+4k}}{2})\right|}
$$

```python
def value_eigen(b, c, k):
    return np.abs(np.maximum(-c, (b+np.sqrt(b**2+4*k))/2))

def norm_F2(a, c):
    return np.maximum(np.abs(a), np.sqrt(c ** 2 +4))

def norm_X0(x0, x1):
    return np.sqrt(x0 ** 2 + x1 ** 2 + x0 ** 4)

def value_R(x0, x1, a, b, c, k):
    return norm_X0(x0, x1) * norm_F2(a, c) / value_eigen(b, c, k)
```

```python
# 定义常数
a = -10
b = -3
k = -1
x0 = 1
x1 = 1

c_values = np.linspace(-10, 40, 2000)

R_values = [value_R(x0, x1, a, b, c, k) for c in c_values]

plt.figure(figsize=(10, 7))
plt.plot(c_values, R_values)
plt.xlabel('c')
plt.ylabel(r'$R = \frac{\|X_{0}\|\max(\left|a\right|,\sqrt{c^{2}+4})}{\left|\max(-c, \frac{b+\sqrt{b^{2}+4k}}{2})\right|}$')
plt.title('a={}, b={}, k={}, x0={}, x1={}'.format(a, b, k, x0, x1))
c_opt = -(b + np.sqrt(b**2 + 4*k))/2
plt.axvline(x=c_opt, linestyle='--', color='gray')

# 添加标注
plt.text(c_opt+0.1, 0.8*np.max(R_values), r'$c=-\frac{b+\sqrt{b^{2}+4k}}{2}=$'+'{:.2f}'.format(c_opt), fontsize=15)
plt.show()
```
