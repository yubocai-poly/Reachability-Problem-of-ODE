```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import solve_ivp
import sympy
from qbee import *
```

$$
\begin{cases}
y' = -2y+2ay^{2}\\
x' = -x + axy
\end{cases}
$$

```python
plt.style.use('seaborn-poster')
%matplotlib inline

a = 5.0
b = 1.0
F = lambda t, y: -2*y + 2*a*y**2
F2 = lambda t, y: -2*y + 2*b*y**2 
```

```python
t_eval = np.linspace(0, 10, 1000)
sol = solve_ivp(F, [0, 10], [0.1], t_eval=t_eval)
sol2 = solve_ivp(F2, [0, 10], [0.1], t_eval=t_eval)

plt.figure(figsize=(10, 6))
plt.plot(sol.t, sol.y[0], 'b', linewidth=2, label='ODE solution 1')
plt.plot(sol2.t, sol2.y[0], 'r', linewidth=2, label='ODE solution 2')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```
